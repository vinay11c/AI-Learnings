**PLATE DETECTION ON IMAGE WITH OCR**

````python
import cv2
import pytesseract
import matplotlib.pyplot as plt
from PIL import Image

# Set the Tesseract path (for Windows)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Load the Haar cascade
plate_cascade = cv2.CascadeClassifier("haarcascade_russian_plate_number.xml")

# Load the image - Update to a valid image path
# Make sure the image file exists and you have permission to access it
image_path = r"C:\Users\VINAY\OneDrive\Pictures\Saved Pictures\download.jpeg"  # Update with correct path inside your Colab environment
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print(f"Error: Could not load image from {r"C:\Users\VINAY\OneDrive\Pictures\Saved Pictures\download.jpeg"}. Please check the file path and permissions.")
else:
    # Proceed with image processing if the image was loaded successfully
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


    # Detect plates
    plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3, minSize=(30, 30))
    # plates = plate_cascade.detectMultiScale(gray, 1.0, 4)

    for (x, y, w, h) in plates:
        # Draw rectangle around the detected plate
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 3)
        plate_region = gray[y:y + h, x:x + w]

        # OCR on the plate region
        plate_text = pytesseract.image_to_string(plate_region, config='--psm 7').strip()
        cv2.putText(image, plate_text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)
        print(f"Detected Plate Text: **{plate_text}**")
    # Show the result
    cv2.imshow('License Plate Detection', image)

    # Wait until a key is pressed to close the window
    cv2.waitKey(0)
    cv2.destroyAllWindows()
````

**PLATE DETECTION ON VDO WITHOUT OCR**

````python
import cv2

# Load the Haar cascade for license plate detection
plate_cascade = cv2.CascadeClassifier("haarcascade_russian_plate_number.xml")

# Load the video
cap = cv2.VideoCapture(r"C:\Users\VINAY\OneDrive\Pictures\Saved Pictures\cars.mp4.mp4")

while cap.isOpened():  # While the video is playing
    has_frame, frame = cap.read()  # Read a frame
    if not has_frame:  # Break if no frame is read
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect plates in the frame
    plates = plate_cascade.detectMultiScale(gray, 1.1, 4)

    # Draw rectangles around detected plates
    for (x, y, w, h) in plates:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)

    # Show the result
    cv2.imshow('Car Number Plate Detection', frame)

    # Exit when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
````

**PLATE DETECTION ON VDO WITH OCR**

````python
import cv2
import pytesseract

# Set the path to the Tesseract executable (if it's not in your PATH environment variable)
# For example, on Windows, the path might look like this:
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Load the Haar cascade for license plate detection
plate_cascade = cv2.CascadeClassifier("haarcascade_russian_plate_number.xml")

# Load the video
cap = cv2.VideoCapture(r"C:\Users\VINAY\OneDrive\Pictures\Saved Pictures\cars.mp4.mp4")

while cap.isOpened():  # While the video is playing
    has_frame, frame = cap.read()  # Read a frame
    if not has_frame:  # Break if no frame is read
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect plates in the frame
    plates = plate_cascade.detectMultiScale(gray, 1.1, 4)

    # Draw rectangles around detected plates
    for (x, y, w, h) in plates:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)

        # Crop the plate region from the frame
        plate_region = frame[y:y + h, x:x + w]

        # Convert the plate region to grayscale for better OCR
        plate_gray = cv2.cvtColor(plate_region, cv2.COLOR_BGR2GRAY)

        # Apply OCR to extract text (license plate number)
        plate_text = pytesseract.image_to_string(plate_gray, config='--psm 8').strip()

        # Display the extracted plate number on the frame
        if plate_text:
            cv2.putText(frame, plate_text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show the result
    cv2.imshow('Car Number Plate Detection', frame)

    # Exit when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
````
